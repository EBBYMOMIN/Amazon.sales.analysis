***4. Exploratory Data Analysis.***

```python
df.columns
```

*size*

```python
ax = sns.countplot(x='Size', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

Note: from above graph we can see that most of the people buy M-Size

*Group By*

```python
df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
S_Qty = df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty', ascending=False)

sns.barplot(x='Size', y='Qty', data=S_Qty)
```

Note: From the above graph we can see that most of the qty buys M_Size in the sale.

*Courier Status*

```python
plt.figure(figsize=(10,5))

sns.countplot(data=df, x='Courier Status', hue='Status')
              
plt.show()
```

Note: From the above graph the majority of the orders are shipped through courier

```python
#histogram

df['Size'].hist()
```

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data, bins=30, edgecolor='Blue')
plt.xticks(rotation=90)
plt.show()
```

Note: From above graph we can se that most of the purchases are T shirt

```python
#Checking B2B data by usinig pie chart
B2B_check = df['B2B'].value_counts()

#Plot the pie chart
plt.pie(B2B_check, labels=B2B_check, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
#Checking B2B data by usinig pie chart
B2B_check = df['B2B'].value_counts()

#Plot the pie chart
plt.pie(B2B_check, labels=B2B_check.index, autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

Note: From the above chart we can see that maximum buyers are retailers 99.2% and 0.8 are B2B buyers

```python
#Prepare data for scatter plot
x_data = df['Category']
y_data = df['Size']

#Plot the scatter plot
plt.scatter(x_data, y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

```python
#Plot count of cities by state

plt.figure(figsize=(20,5))
sns.countplot(data=df, x='ship-state')
plt.xlabel('ship_state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
#top 10 states
top_10_states = df['ship-state'].value_counts().head(10)

#plot count of cities by state
plt.figure(figsize=(15,5))
sns.countplot(data=df[df['ship-state'].isin(top_10_states.index)], x='ship-state')
plt.xlabel('ship_state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

Note: From the above graph we can see that most of the buyers are from Maharashtra state.
